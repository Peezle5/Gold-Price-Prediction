```python
#splitting the features and target
X=g_prices.drop(['Date', 'GLD'], axis=1) #features
y=g_prices['GLD'] #target
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
#splitting the data set into training data and test data
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
#using random forest regression
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor(n_estimators=100)
model.fit(X_train, y_train)
```

```python
y_predicted=model.predict(X_test)
y_predicted
```

```python
#plotting the values of actual prices against the predicted prices to know how close the predictions are to the actual prices.
plt.plot(y_test, color='red', label='Actual Value')
plt.plot(y_predicted, color='blue', label='Predicted Value')
plt.title('Actual Price' and 'Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()
```

```python
#evaluate model
MSE= metrics.mean_squared_error(y_test, y_predicted)
MSE
```

```python
#R squared error(coefficient of determination instead of accuracy score)
r_2=metrics.r2_score(y_test, y_predicted) 
r_2
```
